```python
#Aggregate + Group by
#Read data

store = pd.read_csv("store.csv")
```

```python
store.shape
```

```python
#1 Clean data by renew column name
col_names = list(store.columns)
```

```python
clean_col_names = [col.lower().replace(" ", "_").replace("-","_") for col in col_names]
```

```python
store.columns = clean_col_names
```

```python
store.head()
```

```python
store["sales"].sum()
```

```python
result = store.groupby(["segment", "category"])["sales"].sum().reset_index()
```

```python
store.groupby("segment")["sales"].agg(["sum","mean","std"]).reset_index()
```

```python
result.to_csv("agg_data.csv")
```
